```python
import random
import time
import sys


class KacisOyunu:
    def __init__(self):
        self.envanter = []
        self.kalan_hamle = 30
        self.sifre = str(random.randint(1000, 9999))
        self.ipucu1 = self.sifre[:2]  # ilk 2 hane
        self.ipucu2 = self.sifre[2:]  # son 2 hane
        self.anahtar_alindi = False
        self.kapi_acildi = False
        self.ikinci_oda = False
        self.oyun_baslat()

    def oyun_baslat(self):
        print("Gozlerini actiginda karanlik bir odadasin. Kapi kilitli.")
        self.menu()

    def menu(self):
        if self.kalan_hamle <= 0:
            print("Zamanin doldu. Guvenlik seni yakaladi. Oyunu kaybettin.")
            return

        print(f"\nHamle hakkin: {self.kalan_hamle}")
        print("\nNe yapmak istiyorsun?")
        print("1 - Masayi incele")
        print("2 - Dolabi ac")
        print("3 - Pencereye bak")
        print("4 - Halinin altina bak")
        print("5 - Kapiyi dene")
        print("6 - Envanteri goster")
        secim = input("Secim (1-6): ")
        self.kalan_hamle -= 1

        if secim == "1":
            self.masa()
        elif secim == "2":
            self.dolap()
        elif secim == "3":
            self.pencere()
        elif secim == "4":
            self.hali()
        elif secim == "5":
            self.kapi()
        elif secim == "6":
            self.envanter_goster()
        else:
            print("Gecersiz secim.")
            self.menu()

    def masa(self):
        print("\nMasada sifreli bir kutu var.")
        if self.anahtar_alindi:
            print("Kutuyu zaten actin.")
        else:
            if "ipucu1" in self.envanter and "ipucu2" in self.envanter:
                print(f"Notlara gore sifreyi giriyorsun: {self.sifre}")
                print("Kutu acildi! Anahtar alindi.")
                self.envanter.append("anahtar")
                self.anahtar_alindi = True
            else:
                tahmin = input("Sifreyi girin: ")
                if tahmin == self.sifre:
                    print("Dogru sifre! Anahtari aldin.")
                    self.envanter.append("anahtar")
                    self.anahtar_alindi = True
                else:
                    print("Yanlis sifre.")
                    if tahmin > self.sifre:
                        print("Sifre daha kucuk.")
                    elif tahmin < self.sifre:
                        print("Sifre daha buyuk.")
        self.menu()

    def dolap(self):
        if "ipucu1" not in self.envanter:
            print("Dolabin icinde bir not buldun. Uzerinde iki rakam yaziyor: " + self.ipucu1)
            self.envanter.append("ipucu1")
        else:
            print("Dolapta baska bir sey yok.")
        self.menu()

    def pencere(self):
        print("Pencere civili ve acilamiyor.")
        self.menu()

    def hali(self):
        if "cakI" not in self.envanter:
            print("Halinin altinda bir cakI buldun. Ayrica bir not var: " + self.ipucu2)
            self.envanter.append("cakI")
            self.envanter.append("ipucu2")
        else:
            print("Halinin altinda baska bir sey yok.")
        if "cakI" in self.envanter and "duvar deldi" not in self.envanter:
            print("CakI ile duvardaki taslardan birini kaldiriyorsun ve eski bir anahtar buluyorsun.")
            self.envanter.append("duvar deldi")
            self.envanter.append("anahtar")
        self.menu()

    def kapi(self):
        if "anahtar" in self.envanter:
            if not self.ikinci_oda:
                print("Kapiyi actin! Ikinci odaya geciyorsun.")
                self.ikinci_oda = True
                self.ikinci_odayi_baslat()
            else:
                print("Ikinci odadan da kactin! Oyunu kazandin.")
        else:
            print("Kapi kilitli. Anahtar gerekiyor.")
        self.menu()

    def ikinci_odayi_baslat(self):
        print("Ikinci odada bir masa ve bir kapali kapi var. Bu odayi da kesfetmen gerekiyor.")
        # Buraya yeni mekanikler eklenebilir
        self.menu()

    def envanter_goster(self):
        print("\nEnvanter:")
        for esya in self.envanter:
            print("- " + esya)
        self.menu()


oyun = KacisOyunu()
```
